```python
import pandas as pd
import numpy as np
import matplotlib as mat
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import yfinance as yf
```

```python
# Definindo uma variável com as listas das ações desejadas (Bradesco, Itaú e Santander)
acoes = ['BBDC4.SA', 'ITUB4.SA', 'SANB11.SA']

# Definindo o intervalo de 5 anos de 01/01/2018 até 31/12/2022
start_date = '2018-01-01'
end_date = '2022-12-31'

# Buscando o histórico de preços das ações
df_acoes = yf.download(acoes, start=start_date, end=end_date)['Close']

# Buscar histórico do Ibovespa
ibovespa = yf.download('^BVSP', start=start_date, end=end_date)['Close']

# Compilando os dados 
df = pd.DataFrame(df_acoes)
df['Preço Ibovespa'] = ibovespa

# Exibir o DataFrame com os dados dos preços das ações e Ibovespa
print(df.head(3))
```

```python
# a ideia aqui seria substituir o valor nulo pelo imediatamente anterior
df.isnull().sum()
```

```python
# Tratamento dos dados
# Verificar quais linhas contêm valores nulos
linhas_nulas = df.isnull().any(axis=1)

# Filtrar o DataFrame para exibir apenas as linhas nulas
df_linhas_nulas = df[linhas_nulas]

# Exibir as linhas nulas
print(df_linhas_nulas)
```

```python
# Preencher os valores nulos com os valores imediatamente anteriores
df = df.fillna(method='ffill')
df['2018-01-24':'2018-01-26']
```

```python
df.isnull().sum()
```

```python
# Gráfico das séries
preco_acoes = ['BBDC4.SA','ITUB4.SA','SANB11.SA']
df[preco_acoes].plot(figsize = (15, 5))
```

```python
# Calcular os retornos percentuais para cada ação
retornos_acoes = df.pct_change()

# Renomear as colunas dos retornos
retornos_acoes.columns = ['Retorno ' + col for col in retornos_acoes.columns]

# Adicionar as colunas de retorno ao DataFrame original
df = pd.concat([df, retornos_acoes], axis=1)

# Retirando o primeiro indice da série

df = df.drop(df.index[0])

# Renomear a coluna preço Ibovesspa

df.rename(columns={'Retorno Preço Ibovespa': 'Retorno Ibovespa'}, inplace=True)

# Exibir o DataFrame atualizado
df.head()
```

```python
import pandas as pd
# Buscando o histórico da taxa Selic
cod_selic = '11'
start_date = '01/01/2018'
end_date = '31/12/2022'

# URL da API do Banco Central do Brasil
url_selic = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.' + cod_selic + \
'/dados?formato=json&dataInicial=' + start_date + '&dataFinal=' + end_date

# Faz a requisição GET e carrega os dados em um DataFrame do pandas
df_selic = pd.read_json(url_selic)

# Dividir a série da taxa Selic por 100
df_selic['valor'] = df_selic['valor'].apply(lambda x: x / 100)

df_selic.head()
```

```python
df_selic.info()
```

```python
# Mesclando os dataframes
# Converter as colunas de datas para o formato adequado
df['Data'] = pd.to_datetime(df.index)
df_selic['data'] = pd.to_datetime(df_selic['data'], dayfirst=True)

# Renomear coluna no DataFrame df_selic
df_selic.rename(columns={'valor': 'Taxa Selic'}, inplace=True)

# Realizar a junção dos DataFrames df e df_selic
df_consol = pd.merge(df, df_selic[['data', 'Taxa Selic']], left_on='Data', right_on='data', how='left')

# Remover coluna de datas redundante
df_consol.drop('data', axis=1, inplace=True)

# Exibir o DataFrame resultante
df_consol.head()
```

```python
# copiando o data frame para ajustar os dados
df_ajust = df_consol.copy()

colunas_del = ['BBDC4.SA','ITUB4.SA','SANB11.SA','Preço Ibovespa']  # Substitua pelos nomes das colunas que você deseja excluir

df_ajust.drop(colunas_del, axis=1, inplace=True)
```

```python
# retirando os outliers
colunas_ajust = list(df_ajust)
for names in colunas_ajust[0:]:
    for x in [names]:
        q75,q25 = np.percentile(df_ajust.loc[:,x],[75,25])
        intr_qr = q75-q25

        max = q75+(1.5*intr_qr)
        min = q25-(1.5*intr_qr)

        df_ajust.loc[df_ajust[x] < min,x] = np.nan
        df_ajust.loc[df_ajust[x] > max,x] = np.nan
```

### Análise Preliminar

```python
# Regressão linear para o o prêmio de risco do itau  

import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

# Dados
x = df_ajust['Premio de Mercado']
y = df_ajust['Premio da ITUB4.SA']

# Cálculo da regressão linear utilizndo scipy.stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Cálculo do R²
r_squared = r_value ** 2

# Equação da reta de regressão
equation = f"y = {intercept:.4f} + {slope:.4f}x "

# Gráfico com reta de regressão e R²
sns.jointplot(data=df_ajust, x="Premio de Mercado", y="Premio da ITUB4.SA", kind='reg', height=6)
sns.regplot(x=x, y=y, scatter=False, color='red')
plt.text(x.min() + 0.1 * x.max(), y.max() - 0.1 * y.max(), f"R² = {r_squared:.4f}\n{equation}", fontsize=12)
plt.show()


""""sns.jointplot(data = df_ajust, x = "Premio de Mercado", y = "Premio da ITUB4.SA", kind = 'reg', height=5)"""
```

## Preparando os dados para o Algoritmo de regressão do Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python

# Criando dataframe para as variáveis dependente (y) e independente (x) para o banco Itaú 
import numpy as np
from sklearn.model_selection import train_test_split

# Variável independente (x)
x_mercado = np.array(df_ajust['Premio de Mercado']).reshape(-1, 1)

# Variável dependente (y)
y_itau = np.array(df_ajust['Premio da ITUB4.SA'])

# Dividindo os dados para treino e teste
x_mercado_treino,x_mercado_teste,y_itau_treino,y_itau_teste = train_test_split(x_mercado,y_itau,test_size=0.2,random_state=42)
```

```python
lista_var = [x_mercado_treino, x_mercado_teste, y_itau_treino, y_itau_teste]
for var in lista_var:
    print('Estrutura de', var.shape)
```

```python
# Criando o modelo de Regressão linear 
modelo = LinearRegression()
modelo.fit(x_mercado_treino,y_itau_treino)
```

```python
plt.scatter(x_mercado,y_itau, color = "blue", label = "Dados Reais Históricos")
plt.plot(x_mercado, modelo.predict(x_mercado), color = "red" , label = "Reta de Regressão com as Previsões do Modelo" )
plt.xlabel("Prêmio de Risco de Mercado")
plt.ylabel("Prêmio de Risco do Itaú")
plt.show()
```

```python
score = modelo.score(x_mercado_teste, y_itau_teste)
beta = modelo.coef_

print(f"O Coeficiente R²: {score:.2f}")
print(f"A equação da reta de regressão é: y = {modelo.intercept_:.4f} + {beta[0]:.4f} ")
```
